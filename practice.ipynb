```python
import numpy as np

# Create sample data
arr = np.array([10, 20, 30, 40, 50])

print("Original Array:")
print(arr)

# Basic Operations
print("\nAddition (+5):")
print(arr + 5)

print("\nSubtraction (-5):")
print(arr - 5)

print("\nMultiplication (*2):")
print(arr * 2)

print("\nDivision (/2):")
print(arr / 2)

# Statistical Operations
print("\nSum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))
print("Standard Deviation:", np.std(arr))

# Reshaping
arr2 = arr.reshape(5, 1)
print("\nReshaped Array (5x1):")
print(arr2)

# Indexing and Slicing
print("\nFirst Element:", arr[0])
print("Last Element:", arr[-1])
print("Elements from index 1 to 3:", arr[1:4])

# Create another array
arr3 = np.array([1, 2, 3, 4, 5])

# Element-wise Operations
print("\nElement-wise Addition:")
print(arr + arr3)

print("\nElement-wise Multiplication:")
print(arr * arr3)

# Matrix Operations
matrix = np.array([[1, 2], [3, 4]])

print("\nMatrix:")
print(matrix)

print("\nTranspose:")
print(matrix.T)

print("\nDeterminant:")
print(np.linalg.det(matrix))

print("\nInverse:")
print(np.linalg.inv(matrix))
```

```python
import numpy as np

# Sample data
data = np.array([12, 18, 25, 30, 45, 50, 60, 72, 85, 90])

print("Data:", data)

# 1. Sum
print("Sum:", np.sum(data))

# 2. Mean (Average)
print("Mean:", np.mean(data))

# 3. Median
print("Median:", np.median(data))

# 4. Standard Deviation
print("Standard Deviation:", np.std(data))

# 5. Variance
print("Variance:", np.var(data))

# 6. Maximum
print("Maximum:", np.max(data))

# 7. Minimum
print("Minimum:", np.min(data))

# 8. Range
print("Range:", np.ptp(data))   # Max - Min

# 9. Percentiles
print("25th Percentile:", np.percentile(data, 25))
print("50th Percentile:", np.percentile(data, 50))
print("75th Percentile:", np.percentile(data, 75))

# 10. Cumulative Sum
print("Cumulative Sum:", np.cumsum(data))

# 11. Product of Elements
print("Product:", np.prod(data))

# 12. Index of Maximum and Minimum
print("Index of Maximum:", np.argmax(data))
print("Index of Minimum:", np.argmin(data))

# 13. Unique Values
print("Unique Values:", np.unique(data))
```

```python
import numpy as np

# 3x3 Matrix
matrix = np.array([[10, 20, 30],
                   [40, 50, 60],
                   [70, 80, 90]])

print("Matrix:\n", matrix)

# Row-wise Sum
print("\nRow-wise Sum:", np.sum(matrix, axis=1))

# Column-wise Sum
print("Column-wise Sum:", np.sum(matrix, axis=0))

# Row-wise Mean
print("Row-wise Mean:", np.mean(matrix, axis=1))

# Column-wise Mean
print("Column-wise Mean:", np.mean(matrix, axis=0))

# Maximum in each Row
print("Row-wise Maximum:", np.max(matrix, axis=1))

# Minimum in each Column
print("Column-wise Minimum:", np.min(matrix, axis=0))
```

```python
import pandas as pd

# Sample data
data = [12, 18, 25, 30, 45, 50, 60, 72, 85, 90]

# Create a DataFrame
df = pd.DataFrame(data, columns=["Marks"])

print("DataFrame:")
print(df)

# 1. Sum
print("\nSum:")
print(df["Marks"].sum())

# 2. Mean
print("\nMean:")
print(df["Marks"].mean())

# 3. Median
print("\nMedian:")
print(df["Marks"].median())

# 4. Standard Deviation
print("\nStandard Deviation:")
print(df["Marks"].std())

# 5. Variance
print("\nVariance:")
print(df["Marks"].var())

# 6. Maximum
print("\nMaximum:")
print(df["Marks"].max())

# 7. Minimum
print("\nMinimum:")
print(df["Marks"].min())

# 8. Count
print("\nCount:")
print(df["Marks"].count())

# 9. Describe (Complete Statistical Summary)
print("\nStatistical Summary:")
print(df.describe())

# 10. Cumulative Sum
print("\nCumulative Sum:")
print(df["Marks"].cumsum())

# 11. Unique Values
print("\nUnique Values:")
print(df["Marks"].unique())

# 12. Value Counts
print("\nValue Counts:")
print(df["Marks"].value_counts())

# 13. Largest and Smallest Values
print("\nTop 3 Largest Values:")
print(df["Marks"].nlargest(3))

print("\nTop 3 Smallest Values:")
print(df["Marks"].nsmallest(3))
```

```python
import pandas as pd

# Sample Data
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Math": [85, 78, 92, 70, 88],
    "Science": [90, 75, 95, 72, 85],
    "English": [80, 82, 91, 68, 87]
}

df = pd.DataFrame(data)

print(df)

# Statistical Summary
print("\nSummary:")
print(df.describe())

# Mean of each subject
print("\nAverage Marks:")
print(df.mean(numeric_only=True))

# Maximum marks
print("\nMaximum Marks:")
print(df.max(numeric_only=True))

# Minimum marks
print("\nMinimum Marks:")
print(df.min(numeric_only=True))

# Correlation
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```
